# GeckoTerminal Exploration

Load snapshots from `data/gecko.db` and explore.

Run from the `crypto_lab/` root: `jupyter notebook notebooks/explore.ipynb`

```python
import sqlite3
from pathlib import Path

import pandas as pd

DB = Path('..') / 'data' / 'gecko.db'
con = sqlite3.connect(DB)
df = pd.read_sql_query('SELECT * FROM pool_snapshot', con)
print(f'{len(df)} rows across {df.network.nunique()} networks')
df.head()
```

```python
# How many unique pools have you seen per chain?
df.groupby('network')['pool_address'].nunique().sort_values(ascending=False)
```

```python
# Pools that first appeared as 'new' and are still alive (>$10K volume 24h)
alive = df[(df.kind == 'new') & (df.volume_h24_usd > 10_000)]
alive.sort_values('volume_h24_usd', ascending=False).head(20)
```
